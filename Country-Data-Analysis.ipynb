# Step 4: Social Welfare Measure

I am normalizing values from the following and using it to predict General Social Welfare.
1. Population
2. GDP
3. Literacy Rate

```python
lifemax=df['Life Expectancy'].max()
gdpmax=df['GDP (PPP) in US $'].max()
#literacy=df.sort('Literacy Rate (%)')
#literacymax=literacy['Literacy Rate (%)'][0]

x=list()
y=list()
z=list()
avg=list()

n=0;
#print('Life expectancy')
for i in df['Life Expectancy']:
    x.append((i*100)/lifemax)
#print x
    
    
#print('GDP')
for i in df['GDP (PPP) in US $']:
    y.append((i*100)/gdpmax)
#print y

#print('Literacy Rate') 
for i in df['Literacy Rate (%)']:
    z.append(i)
#print z

for i in x:
    avg.append((x[n]+y[n])/2)
    n=n+1;
#print avg
```

```python
%matplotlib inline
```

```python
import matplotlib.pylab as plt
cname = df.index.values
y_pos = np.arange(len(cname))
plt.figure(figsize=(180, 40), dpi=20)
plt.bar(y_pos, avg, align='edge')
plt.yticks(y_pos, cname)
plt.xlabel('Welfare Measure')
plt.title('General Social Welfare Measure')
#plt.show()
plt.savefig('SocialWelfare.png', bbox_inches='tight')
```

It appears that the above is a power law distribution. Two countries seem to do exceptionally well on the general social welfare, while the others are pretty much in the same range
